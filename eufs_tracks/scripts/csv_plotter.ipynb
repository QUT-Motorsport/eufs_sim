```python
import pandas as pd

# Load the CSV file
file_path = '../csv/small_track_2.csv'
df = pd.read_csv(file_path)

# Display the first few rows of the dataframe to understand its structure
df.head()
```

```python
%matplotlib qt
import matplotlib.pyplot as plt

# Assigning colors to each tag
color_map = {
    'blue': 'blue',
    'yellow': 'gold',
    'big_orange': 'orange',
    'car_start': 'red'
}

# Plotting the data
plt.figure(figsize=(10, 6))

for tag, group in df.groupby('tag'):
    plt.scatter(group['x'], group['y'], label=tag, color=color_map.get(tag, 'black'))

plt.ion()
plt.xlabel('X Coordinate')
plt.ylabel('Y Coordinate')
plt.title('Visualization of Cones from CSV Data')
plt.legend()
plt.grid(True)
plt.show()
```
